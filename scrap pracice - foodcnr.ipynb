```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd 
import os

df = pd.read_csv("foodcorner_links.csv")

links = df["link"].values.tolist()


for link in links:
    print(link)
    
    # Send an HTTP GET request to the webpage
    url = link
    response = requests.get(url)

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the recipe title
    recipe_title = soup.find("h1", class_ = "post-title single-post-title")
    if recipe_title:
        recipe_title = recipe_title.text.strip()
        print(recipe_title)
        
    # Find the ingredients list
    ingredients = soup.find_all("li", class_="wprm-recipe-ingredient")
    ingredient_list = [ingredient.text.strip() for ingredient in ingredients]

    # Find the instructions
    instructions = soup.find_all("li", class_="wprm-recipe-instruction")
    instruction_list = []
    for index, instruction in enumerate(instructions, 1):
        instruction_text = instruction.text.strip()
        instruction_list.append({"number": index, "instruction": instruction_text})

    # Create a dictionary with the recipe details
    recipe_data = {
        "title": recipe_title,
        "ingredients": ingredient_list,
        "instructions": instruction_list
    }

    filename = f"{recipe_title.replace(' ', '_')}.json"
    filepath = os.path.join("recipes", filename)
    
    # Save the recipe data as JSON
    with open( filepath, 'w',encoding='utf-8') as json_file:
        json.dump(recipe_data, json_file, indent=4)
```

```python
import requests
from bs4 import BeautifulSoup

# List of category links
category_links = [
    "https://foodcnr.com/category/air-fryer-recipes/",
    "https://foodcnr.com/category/appetizers-snacks/",
    "https://foodcnr.com/category/bread-bun-pizza-baking-recipes/"
]

# Iterate over the category links
for category_link in category_links:
    # Make a request to the category page
    response = requests.get(category_link)

    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the pagination links
    pagination_links = soup.find_all('a', class_='page-numbers')
    page_links = [category_link] + [link['href'] for link in pagination_links]

    # Iterate over the page links within the category
    for page_link in page_links:
        # Make a request to the page
        response = requests.get(page_link)

        # Parse the HTML content
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all the recipe links on the page
        recipe_links = soup.find_all('a', class_='entry-title-link')

        # Iterate over the recipe links
        for recipe_link in recipe_links:
            # Make a request to the recipe page
            response = requests.get(recipe_link['href'])

            # Parse the HTML content
            soup = BeautifulSoup(response.content, 'html.parser')

            # Scrape Ingredients
            ingredient_heading = soup.find('h2', text='Ingredients')
            if ingredient_heading:
                ingredients = []
                for sibling in ingredient_heading.find_next_siblings():
                    if sibling.name == 'h2':
                        # Stop when encountering the next heading
                        break
                    ingredients.append(sibling.get_text().strip())
                print("Ingredients:")
                for ingredient in ingredients:
                    print(ingredient)

            # Scrape Instructions
            instructions_heading = soup.find('h2', text='Instructions')
            if instructions_heading:
                instructions = []
                for sibling in instructions_heading.find_next_siblings():
                    if sibling.name == 'h2':
                        # Stop when encountering the next heading
                        break
                    instructions.append(sibling.get_text().strip())
                print("\nInstructions:")
                for instruction in instructions:
                    print(instruction)
            
            print("\n--- End of Recipe ---\n")

        print("\n--- End of Page ---\n")

    print("\n--- End of Category ---\n")
```
